### SAVE RAW DATA TO JSON

```python
import json
with open('yelp_data_raw.json', 'w') as fp:
    json.dump(json_data, fp)
```

```python
# pprint(json_data)
```

### STORE RAW JSON DATA TO DATA FRAME
#### For importing to CSV

```python
import pandas as pd
df = pd.DataFrame.from_dict(json_data)
df.head()
```

```python
df.shape
```

```python
cols = list(df.columns.values)
cols
```

### Rearrange columns

```python
df_ordered = df [['name', 'location', 'rating', 'review_count', 'price', 'categories', 'coordinates']]
df_ordered.head()
```

### Add lat ang long columns (for extra options)

```python
df_yelp = df_ordered.copy()

lat_list = [d.get('latitude') for d in df_ordered.coordinates]
lon_list = [d.get('longitude') for d in df_ordered.coordinates]

df_yelp['latitude'] = lat_list
df_yelp['longitude'] = lon_list
df_yelp.head()
```

```python
df_yelp.to_csv("yelp_data_raw.csv");
```

### Clean The data

```python
df_yelp_for_cleaning = df_yelp.copy()
df_yelp_for_cleaning_no_none = df_yelp_for_cleaning.dropna()
df_yelp_for_cleaning_no_none.head()
```

```python
df_yelp_for_cleaning_no_none.shape
```

### Remove duplicates

```python
df_yelp_no_dupes = df_yelp_for_cleaning_no_none.copy()
df_yelp_no_dupes.drop_duplicates(subset =["name", "categories"], 
                     keep = "first", inplace = True) 
df_yelp_no_dupes
```

```python
df_yelp_no_dupes.shape
```

### There can be multiple instances of a restaurant that varies in its categories

```python
print(df_yelp_no_dupes[df_yelp_no_dupes['name']=='San Tung'])
```

### Save clean data to csv

```python
df_yelp_clean_data =  df_yelp_no_dupes.copy()
df_yelp_clean_data.to_csv("yelp_data_clean.csv", index=False);
```

```python
df_yelp_clean_data.head()
```

### Save clean data to Json file

```python
json_data_clean = []
for index, row in df_yelp_clean_data.iterrows():
    t_dictionary = {"name":row["name"],
                    "categories":row["categories"],
                    "coordinates":row["coordinates"],
                    "location":row["location"],
                    "price":row["price"],
                    "rating":row["rating"],
                    "review_count":row["review_count"],
                    "latitude":row["latitude"],
                    "longitude":row["longitude"]
                   }
    json_data_clean.append(t_dictionary) 
#     print(json_data_clean)
```

```python
import json
with open('yelp_data_clean.json', 'w') as fp:
    json.dump(json_data_clean, fp)
print("done!")
```

```python
df_yelp_clean_data.columns.tolist()
```

### Load CSV file

```python
import pandas as pd
file_name = "yelp_data_clean.csv"
loaded_yelp = pd.read_csv(file_name)
print(loaded_yelp.iloc[[0]].name)
loaded_yelp.head()
```

```python
loaded_yelp.shape
```

## Save clean dataframe to sqlite

```python
from sqlalchemy import create_engine
engine = create_engine('sqlite:///yelp_data.sqlite', echo=False)
```

```python
loaded_yelp.name = loaded_yelp['name'].astype(str)
loaded_yelp.dtypes
```

```python
loaded_yelp.to_sql('yelp_sf', con=engine, if_exists='replace', index=False)
```

```python
from sqlalchemy import inspect

inspector = inspect(engine)
inspector.get_table_names()
```

```python
columns = inspector.get_columns('yelp_sf')
for c in columns:
    print(c['name'], c["type"])
```

```python
engine.execute('SELECT * FROM yelp_sf LIMIT 10').fetchall()
```

## Save clean Json file to MongoDB

```python
import pymongo
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
yelp_db = client.yelp
yelp_collection = yelp_db.yelp_collection.find()

#Insert Data
path = 'yelp_data_clean.json'
with open(path) as json_file:
    data = json.load(json_file)

yelp_db.yelp_collection.remove()
for d in data:
    yelp_db.yelp_collection.insert_one(d)
 
```

```python
# # Query data
yelp_mongo = yelp_db.yelp_collection.find()
for row in yelp_mongo:
    print(row)
```
